```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/nba_season_player_stats.csv')
```

```python
mask = (df['pts'] >= 10) & (df['gp'] >= 20)
x = df.loc[mask, 'fg_pct']
y = df.loc[mask, 'pts'] / df.loc[mask, 'gp']
plt.scatter(x, y)
plt.xlabel('Field Goal %')
plt.ylabel('Points Per Game')
plt.title("FG% vs PPG")
plt.grid(True)
plt.show()
```

```python
df['ppg'] = df['pts'] / df['gp']
print(df[df['ppg'] > 50][['player_id', 'season_id', 'pts', 'gp', 'ppg']])
```
